```python
import json
import folium

map = folium.Map([37.487935, 126.857758], zoom_start=11)
with open('E:/21KDT-Project-master/spatial_analysis-master/data/TL_SCCO_CTPRVN.json', encoding='UTF-8') as f:
    data = json.load(f)
    
incheon = data['features'][11]

map.choropleth(
    geo_data=incheon,
    line_color="green", 
    line_weight=5, 
    #fill_color='white', 
    fill_color='YlGn',
    fill_opacity=0.1, 
    line_opacity=5,
    smooth_factor=0)

map
```
